```python
import pandas as pd
```

```python
df=pd.read_csv('glassdoor_jobs.csv')
```

```python
df.head(5)
```

# Data Cleaning task list
1.salary parsing<br>
2.company name text only<br>
3.company age<br>
4.state field only<br>
5.jobs description parsing<br>

```python
#task 1

df['per_hour'] = df['Salary Estimate'].apply(lambda x:1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x:1 if 'employer provided salary:' in x.lower() else 0)
df =  df[df['Salary Estimate']!='-1']
df['salary']= df['Salary Estimate'].apply(lambda x: x.split('(')[0])

#removing $ and K
df.salary=df.salary.apply(lambda x :x.replace('$',' ').replace('K',' '))
df.salary = df.salary.apply(lambda x :x.replace('Per Hour',' ').replace('Employer Provided Salary:',' '))

#min, max and avg salary
df['min_salary'] = df.salary.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = df.salary.apply(lambda x: int(x.split('-')[1]))
df['avg_salary'] = (df.min_salary+df.max_salary)/2
```

```python
#task 2
df['company_name'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-5], axis=1 )
```

```python
#task 3
df['company_age'] = df['Founded'].apply(lambda x: x if x<1 else 2020-x)
```

```python
#task 4
df['jobs_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0,axis=1)
```

```python
#task 5 
df['python_ys'] =df['Job Description'].apply(lambda x:1 if 'python' in x.lower() else 0)
df['sql_ys']    =df['Job Description'].apply(lambda x:1 if 'sql' in x.lower() else 0)
df['excel_ys']  =df['Job Description'].apply(lambda x:1 if 'excel' in x.lower() else 0)
df['spark_ys']  =df['Job Description'].apply(lambda x:1 if 'spark' in x.lower() else 0)
df['r_ys']      =df['Job Description'].apply(lambda x:1 if 'r studio' in x.lower() else 0)
```
